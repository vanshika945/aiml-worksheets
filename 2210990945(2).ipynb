# DISPLAY ALL THE EMPLOYEE NAMES FROM FIRE DEPARTMENT

# FIND MINIMUM,MAXIMUM AND AVERGE BASEPAY

```python
import pandas as pd


df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')

min_basepay = df['BasePay'].min()


max_basepay = df['BasePay'].max()


avg_basepay = df['BasePay'].mean()

print( min_basepay)
print( max_basepay)
print( avg_basepay)
```

# REPLACE 'NOT PROVIDED' IN  EMPLOYEENAME COLUMN TO NaN

```python
import pandas as pd
import numpy as np

df['EmployeeName'] = df['EmployeeName'].replace('NOT PROVIDED', np.nan)
print(df)
```

# drop the rows having more than 5 missing values

```python
import pandas as pd
threshold = len(df.columns) - 5 
df = df.dropna(thresh=threshold)
print(df)
```

# FIND AVERGE BASEPAY OF ALL EMPLOYEE PER YEAR

```python
import pandas as pd
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')


average_basepay_per_year = df.groupby('Year')['BasePay'].mean()

print(average_basepay_per_year)
```

# FIND AVERGE BASEPAY OF ALL EMPLOYEE PER JOB TITLE

```python
import pandas as pd
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')

average_basepay_per_job_title = df.groupby('JobTitle')['BasePay'].mean()

print(average_basepay_per_job_title)
```

# FIND AVERGE BASEPAY OF EMPLOYEE HAVIG JOB TITLE ACCOUNTANT

```python
import pandas as pd

df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')

accountant_data = df[df['JobTitle'] == 'ACCOUNTANT']

average_basepay_accountant = accountant_data['BasePay'].mean()

print( average_basepay_accountant)
```
