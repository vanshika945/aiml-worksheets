# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
import matplotlib.pyplot as plt
survived_count=data['Survived'].value_counts()

#plot the counts
plt.figure(figsize=(8,6))
plt.bar(survived_count.index,survived_count.values,color=['red','green'])
plt.xticks(survived_count.index,['Diet','Survived'])
plt.xlabel('Outcome')
plt.ylabel('Number of people')
plt.title('Number of people survived and diet')
plt.show()
```

# HOW MANY PASSENGERS WERE IN FIRST CLASS,SECOND CLASS AND THIRD CLASS PLOT THOSE FIGURES ON GRAPH

```python
class_count=df['Pclass'].value_counts()

#plot the pie chart
plt.figure(figsize=(8,6))
plt.pie(class_count,labels=class_count.index,autopct='%1.1f%%',colors=['blue','green','red'])
plt.title('Distribution of passengers by class')
plt.legend(labels=['1st class','2nd class','3rd class'],loc='upper right')

plt.show()
```

# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
survival_pivot = data.pivot_table(index='Sex', columns='Survived', aggfunc='size', fill_value=0)

# Plot the survival rates
survival_pivot.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'green'])
plt.xlabel('Sex')
plt.ylabel('Count'),
plt.title('Survival Count by Sex')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.xticks(rotation=0)
plt.show()
```

# WHICH PASSENGER CLASS HAS BETTER CHANCE OF SURVIVAL(FIRST,SECOND,OR THIRD CLASS)?

```python
survival_pivot=data.pivot_table(index='Pclass',columns='Survived',aggfunc='size',fill_value=0)

survival_pivot.plot(kind='bar',stacked=True,figsize=(8,6),color=['yellow','blue'])
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Survival Count by Passenger Class ')
plt.legend(title='Survived',labels=['No','Yes'])
plt.show()
```
